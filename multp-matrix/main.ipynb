```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfs = [
    pd.read_csv('./10x10.csv'),
    pd.read_csv('./10-2x10-2.csv'),
    pd.read_csv('./10-3x10-3.csv'),
    pd.read_csv('./10-4x10-4.csv'),
]
```

```python
means_cpu = []
means_gpu = []

for i in dfs:
    means_cpu.append(i['CPU'].mean())
    means_gpu.append(i['GPU'].mean())
```

```python
fig, axes = plt.subplots()

width = 0.40
x = np.arange(4)

axes.bar(x-width/2, means_cpu, width=width, label="CPU")
axes.bar(x+width/2, means_gpu, width=width, label="GPU")

axes.set_xticks(x, ['10x10', '100x100', '1000x1000', '10000x10000']) 
axes.set_xlabel("Ukuran") 
axes.set_ylabel("Waktu (Detik)")
axes.set_title('Operasi Penjumlahan Matriks')
axes.set_yscale('log')

axes.legend()
axes.grid()
fig.tight_layout()
```
