```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dfs = [
    pd.read_csv('./1.csv'),
    pd.read_csv('./2.csv'),
    pd.read_csv('./3.csv'),
    pd.read_csv('./4.csv'),
    pd.read_csv('./5.csv'),
    pd.read_csv('./6.csv'),
    pd.read_csv('./7.csv'),
]
```

```python
means_cpu = []
means_gpu = []

for i in dfs:
    means_cpu.append(i['CPU'].mean())
    means_gpu.append(i['GPU'].mean())
```

```python
means_cpu
```

```python
means_gpu
```

```python
df_mean = pd.DataFrame()
df_mean['means_cpu'] = means_cpu
df_mean['means_gpu'] = means_gpu
df_mean
```

```python
len(dfs)
```

```python
TITLE = 'Operasi Pengurangan Matriks'
fig, axes = plt.subplots(figsize=(12, 8))

width = 0.40
x = np.arange(len(dfs))

axes.bar(x-width/2, means_cpu, width=width, label="CPU")
axes.bar(x+width/2, means_gpu, width=width, label="GPU")

axes.set_xticks(x, ['10x10', '50x50', r'$10^2$x$10^2$', '500x500', '$10^3$x$10^3$', '5000x5000', '$10^4$x$10^4$']) 
axes.set_xlabel("Ukuran") 
axes.set_ylabel("Waktu (Detik)")
axes.set_title(TITLE)
axes.set_yscale('log')

axes.legend()
axes.grid()
# fig.tight_layout()
```
